```python
%matplotlib widget
import pandas as pd
import json
from flatten_json import flatten
import math
```

```python
def parsePlanks(planks):
    return [p["Wheel"] for p in planks]

def parsePassengers(passengers):
    return [[p["Head"], p["Body"]] for p in passengers]

def parseWagons(wagons):
    return [[parsePlanks(w["Wagon"]["Planks"])] + [parsePassengers(w["Passengers"])] for w in wagons]

def parseLocomotive(locomotive):
    return parsePlanks(locomotive["Planks"])
    
def parseTrain(train):
    return [[parseLocomotive(train["Locomotive"])] + w  for w in parseWagons(train["Wagons"])]

def parseTrains(trains):
    x = [parseTrain(t) for t in trains]
    return [[i, j] + w for (i, t) in enumerate(x) for (j, w) in enumerate(t)]

def parseResult(result):
    return [[result["X"]] + r for r in parseTrains(result["Trains"])]

def parseResults(results):
    return [parseResult(r) for r in results]
    
```

```python
with open('../data/res.json') as data_file:
    data = json.load(data_file)

names = ["X", "TrainId", "WagonId", "LocomotiveWheels", "WagonWheels", "WagonPassengers"]
  
listOfLists = [y for res in parseResults(data["Results"]) for y in res]
trains = pd.DataFrame(listOfLists, columns=names)
trains['#LocomotivePlanks'] = trains['LocomotiveWheels'].apply(len)
trains['#WagonPlanks'] = trains['WagonWheels'].apply(len)
trains['#WagonPassengers'] = trains['WagonPassengers'].apply(len)
trains['#WagonWheels'] = trains['WagonWheels'].apply(lambda wheels: sum(1 for num in wheels if num != 0))
trains['#WagonWheelsSum'] = trains['WagonWheels'].apply(sum)
trains['#WagonWheelsSum'] = trains['WagonWheels'].apply(sum)
trains['#WagonPlanks'] = trains['WagonWheels'].apply(len)
trains['#LocomotiveWheels'] = trains['LocomotiveWheels'].apply(lambda wheels: sum(1 for num in wheels if num != 0))
trains['#LocomotiveWheelsSum'] = trains['LocomotiveWheels'].apply(sum)
Xs = trains.X.unique()

trains.head(5)  



# trains.loc[trains.X == trains.X.unique()[0]].groupby(['TrainId'])[].head()
```

```python
def chebyshev(data, percent):
        EX = sum([abs(dat) for dat in data]) / (len(data)- 1)
        VarX = sum([(dat - EX) ** 2 for dat in data]) / (len(data) - 1)
        return math.sqrt(VarX/ percent)

def markov(data, percent):
        EX = sum([abs(dat) for dat in data]) / (len(data)- 1)
        return EX / percent

def getUpperAndLower(data, percent):
    sortedData = sorted(data)
    return sortedData[int(percent * len(data))], sortedData[-(int(percent * len(data)) + 1)]
```

```python
import matplotlib.pyplot as plt

def create_scatter(label):
    x_values = trains['X']
    y_values = trains[label]

    # Create a scatter plot
    plt.scatter(x_values, y_values)

    # Set labels for axes
    plt.xlabel('X')
    plt.ylabel(label)

    # Set a title for the plot
    plt.title(f"Scatter Plot of {label} with regards to X")

    # Show the plot
    plt.show()
```

```python
def create_summary_plot_wagons(label):
    grouped_df = trains.groupby(['X', 'NoOfTrain'])['WheelCount'].count().reset_index()


    summary_stats = grouped_df.groupby('X')['WheelCount'].agg(['min', 'max', 'mean']).reset_index()

    # Plotting
    plt.plot(summary_stats['X'], summary_stats['min'], label='Min')
    plt.plot(summary_stats['X'], summary_stats['max'], label='Max')
    plt.plot(summary_stats['X'], summary_stats['mean'], label='Average')

    # Set labels for axes
    plt.xlabel('X')
    plt.ylabel(label)

    # Set a title for the plot
    plt.title(f"Min, Max, and Average {label} for each X")

    # Add a legend
    plt.legend()

    # Show the plot
    plt.savefig(f'results/summary_{label}.png')
    plt.show()
    
```

```python
plt.close('all')
fig, axs = plt.subplots(3, figsize=(8, 24))
create_summary_plot_wagons('Wagons')
```

```python
def PlotConentrationForAll(label, agg):
    train_wagon_counts = trains.groupby(['X', 'TrainId'])[label].apply(agg).reset_index()
    print(train_wagon_counts.head())
# Calculate the mean number of wagons per train for each test
    mean_wagons_per_train = train_wagon_counts.groupby('X')[label].mean().reset_index()[label]

# Calculate Chebyshev scores for each test using your custom function
    chebyshev_scores = trains.groupby('X')[label].apply(lambda x: chebyshev(x, 0.2)).values
    mininmum = trains.groupby('X')[label].apply(min).values
    maximum = trains.groupby('X')[label].apply(max).values

    std = trains.groupby('X')[label].std().values

    chebyshev_up = [m + c for m, c in zip(mean_wagons_per_train, chebyshev_scores)]
    chebyshev_down = [m - c for m, c in zip(mean_wagons_per_train, chebyshev_scores)]
    std_up = [m + 2 * s for m, s in zip(mean_wagons_per_train, std)]
    std_down = [m + 2 * s for m, s in zip(mean_wagons_per_train, std)]

# Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(trains.X.unique(), chebyshev_up, label='Cheby up')
    plt.plot(trains.X.unique(), chebyshev_down, label='Cheby down')
    plt.plot(trains.X.unique(), std_up, label='std up')
    plt.plot(trains.X.unique(), std_down, label='std down')
    plt.plot(trains.X.unique(), mean_wagons_per_train, label='avg')
    plt.plot(trains.X.unique(), mininmum, label='Min')
    plt.plot(trains.X.unique(), maximum, label='Max')
    plt.xlabel('Test ID')
    plt.ylabel(label)
    plt.xticks(rotation=60)
    plt.title(f'Koncentracja {label}')
    plt.legend()
    plt.show()

# PlotConentrationForAll("WagonId", lambda x: x.count())
# PlotConentrationForAll("#WagonWheelsSum", lambda x: x.mean())
# PlotConentrationForAll("#LocomotiveWheelsSum", lambda x: x.mean())
# PlotConentrationForAll("#WagonWheels", lambda x: x.mean())
# PlotConentrationForAll("#LocomotiveWheels", lambda x: x.mean())
# PlotConentrationForAll("#WagonPassengers", lambda x: x.mean())
```

```python
import numpy as np

def plotHistogram(label, agg):
    # Assuming df is your DataFrame
    # Calculate the number of wagons in each train for all Xs
    wagon_counts_all = trains.groupby(['X', 'TrainId'])[label].apply(agg)

    # Create a 3D surface plot
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Get unique Xs
    unique_test_ids = trains['X'].unique()

    # Set up variables for 3D surface plot
    hist_data = []
    x_bins = len(unique_test_ids)
    y_bins = 40  # Adjust the number of bins for the number of wagons as needed
    x_range = Xs
    y_range = range(y_bins)

    # Populate the histogram data
    for test_id in unique_test_ids:
        wagon_counts = wagon_counts_all[test_id].values
        hist, _ = np.histogram(wagon_counts, bins=y_bins, range=(0, y_bins))
        hist_data.append(hist)

    # Convert the histogram data to a 2D array
    hist_data = np.array(hist_data)

    # Create X, Y grids
    X, Y = np.meshgrid(x_range, y_range)

    # Plot the 3D surface plot
    surf = ax.plot_surface(X, Y, hist_data.T, cmap='viridis')

    # Customize the labels
    ax.set_xlabel('X')
    ax.set_ylabel(f'{label} in Train')
    ax.set_zlabel('Frequency')

    # Add a colorbar
    fig.colorbar(surf)

    plt.show()
```

```python
plotHistogram("WagonId", lambda x: x.count())
plotHistogram("#WagonPassengers", lambda x: x.sum())
plotHistogram("#WagonWheels", lambda x: x.sum())
plotHistogram("#WagonPlanks", lambda x: x.sum())
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plotAll(label):
    # Aggregate data
    aggregated_data = trains.groupby(['X', 'TrainId'])[label].sum().reset_index()

    # Pivot the DataFrame to get a matrix form
    pivot_table = aggregated_data.pivot(index='X', columns='TrainId', values=label).fillna(0)

    # Create meshgrid
    X, Y = np.meshgrid(pivot_table.columns, pivot_table.index)
    Z = pivot_table.values
    # Plotting
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')

    ax.set_xlabel('TrainId')
    ax.set_ylabel('X')
    ax.set_zlabel('Sum of ' + label)
    ax.set_zscale('log')

    plt.show()

def plot_2d_histogram_at_max_x(label):
    max_x = trains['X'].max()
    print(max_x)
    max_x_data = trains[trains['X'] == max_x]
    # print(max_x_data.head(5))
    # Sum the property for each TrainId at max X
    print(max_x_data.groupby('TrainId').get_group(0))
    sums_at_max_x = max_x_data.groupby('TrainId')[label].sum()

    # Plotting the histogram
    plt.figure(figsize=(8, 6))
    plt.hist(sums_at_max_x, bins=60, alpha=0.7)
    plt.title('Histogram of Sum of ' + label + ' for each TrainId at Max X')
    plt.xlabel('Sum of ' + label)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.close('all')
plotAll("#WagonPassengers")
plot_2d_histogram_at_max_x("#WagonPassengers")
```
